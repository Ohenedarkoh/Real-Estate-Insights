```python
# Split amenities into individual columns
amenities =housing['Amenities'].str.split(', ', expand=True)

# Get unique amenities and create binary columns
all_amenities = set()
for i in range(amenities.shape[1]):
    all_amenities.update(amenities[i].dropna().unique())

for amenity in all_amenities:
    if pd.notna(amenity):  # Skip NaN values
        housing[f'Amenity_{amenity.strip()}'] = housing['Amenities'].str.contains(amenity, regex=False)
```

```python
housing['Reference'] = housing['Reference'].astype(str).str.strip()
housing

#clearig whitespaces and converting column to string
```

```python
# Split amenities into individual columns
amenities = housing['Amenities'].str.split(', ', expand=True)

# Get unique amenities and create binary columns
all_amenities = set()
for i in range(amenities.shape[1]):
    all_amenities.update(amenities[i].dropna().unique())

for amenity in all_amenities:
    if pd.notna(amenity):  # Skip NaN values
        housing[f'Amenity_{amenity.strip()}'] = housing['Amenities'].fillna('').str.contains(amenity, regex=False).astype(int)



# Drop original column
# housing = housing.drop('Amenities', axis=1) not neccesary!
```
